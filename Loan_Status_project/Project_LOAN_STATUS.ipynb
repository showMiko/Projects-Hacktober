```python
 loan_dataset.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'Yes':1,'No':0}, 'Education':{'Graduate':1,'Not Graduate':0},'Property_Area':{'Rural':0,'Urban':1,'Semirural':2,'Semiurban':3}},inplace=True)
```

```python
loan_dataset.head()
```

```python
x=loan_dataset.drop(columns=['Loan_ID'],axis=1) 
y=loan_dataset['Loan_Status']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.1,stratify=y ,random_state=2)
```

```python
loan_dataset.head()
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(x_train,y_train)
```

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print('Accuracy=', training_data_accuracy)
```

```python
x_test_prediction = classifier.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print('Accuracy=', test_data_accuracy)
```
